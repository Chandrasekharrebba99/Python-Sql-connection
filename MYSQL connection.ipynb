# Convert Mysql DataTable to Pandas DataFrame

```python
import mysql.connector as connection #importing Mysql connector 
import pandas as pd #importing Pandas library
try:
    mydb = connection.connect(host = "localhost",database = "sql_store",username = "root",password = "Chandu@123",use_pure=True)
    print(mydb.is_connected()) #print True if it connected
    query = "Select * from customers"  #query go's here
    cursor = mydb.cursor() #create cursor
    cursor.execute(query) #run cursor 
    
    df = pd.DataFrame(cursor.fetchall()) #converting sql Data to pandas Dataframe
    print(df)
except Exception as e:
    mydb.close()
    print(str(e))
```
